```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy.io import arff
```

```python
arff_file = arff.loadarff('Training Dataset.arff')

# Convert the dataset to a Pandas DataFrame
phishing = pd.DataFrame(arff_file[0])

phishing

# Print out the unique values in each column
#for col in phishing.columns:
    #print(col, phishing[col].unique())
```

```python
# Check for duplicated rows
duplicated_rows = phishing.duplicated()

# Count the number of duplicated rows
num_duplicates = duplicated_rows.sum()

# Print out the number of duplicated rows
print(f"Number of duplicated rows: {num_duplicates}")
```

```python
phishing = phishing.drop_duplicates()

# Print out the shape of the DataFrame to confirm the removal of duplicated rows
print(f"Shape of DataFrame after dropping duplicates: {phishing.shape}")
```

```python
# Check for missing values
missing_values = phishing.isnull().sum()

# Print out the missing values
print(missing_values)
```

```python
# there are no missing values in dataset
```

```python
# Visualize the distribution of each column using histograms
for column in phishing.columns:
    if phishing[column].dtype in [float, int]:
        phishing[column].plot(kind='hist', bins=10, figsize=(5,3))
        plt.title(column)
        plt.show()
    elif phishing[column].dtype == 'datetime64[ns]':
        phishing[column].hist(bins=10, figsize=(5,3))
        plt.title(column)
        plt.show()
    else:
        print(f"Skipping {column} because it is not numerical or datetime")
```
